Let's create some artifical data:

```python
# Create numpy array with 20 evenly spaced numbers over interval 0 to 20.
x = linspace(0, 20, 20)
# Each element to the power of 5.
y = x ** 5
```

Plotting is as easy as this:

```python
plot(x, y)
```

Plotting styles and labels are controlled in a way similar to Matlab:

```python
plot(x, y, 'r--o')
xlabel('x')
ylabel('y')
title('My first nice plot with dots.')
```

You can find the plot() documentation here:

Make it more fancy, like xkcd plots! http://xkcd.com/1299/
Hint: For the xkcd style you need a recent matplotlib version, but it's worth it.

```python
with xkcd():
    plot(x, y, 'r--o')
    xlabel('x')
    ylabel('y')
    title('My first nice plot with dots.')
```

You can plot several individual lines at once:

```python
plot(x, y, 'r--o', x, y ** 1.1, 'bs', x, y ** 1.2, 'g^-' )
```

The following example is from [matplotlib - 2D and 3D plotting in Python](http://nbviewer.ipython.org/urls/raw.github.com/jrjohansson/scientific-python-lectures/master/Lecture-4-Matplotlib.ipynb) - a great place to start for people interested in matplotlib.


```python
n = array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
x = linspace(0, 5, 10)

fig, axes = plt.subplots(1, 4, figsize=(12,3))
axes[0].scatter(xx, xx + 0.25*randn(len(xx)))
axes[1].step(n, n**2, lw=2)
axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5)
axes[3].fill_between(x, x**2, x**3, color="green", alpha=0.5)
```

After having sequenced a genome and having obtained a high quality resolution of its DNA sequence,
one of the initial tasks is to predict and annotate the protein-coding genes.

Protein-coding regions of a DNA sequence are first transcribed into messenger RNA (mRNA) and
then translated into a protein. During this process, a codon (consisting of 3 nucleotides) is translated
into a single amino acid and joined together to form a polypeptide, i.e. a protein. The mRNA contains
signals that tell the ribosome where to initiate and terminate translation, the sequence between
these two signals is the coding sequence (CDS), often also called open reading frame (ORF). 
Such an open reading frame is a sequence that contains a start and a stop codon and is readable by the ribsome. 
A single-stranded DNA sequence contains three reading frames, this means three possibilities of 
reading codons off the sequence, starting from position one, two, or three. The codons are then 
just read as a consecutive sequence of three nucleotides.

$$S[i], S[i + 1], . . . , S[j]$$
